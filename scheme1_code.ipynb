```python
import random
def random_salt(length):
  result=''
  for i in range(0,length):
    result+=str(random.randint(0,9))

  return result
```

```python
df_1['salt'] = df_1['password'].apply(lambda password: random_salt(3))
```

```python
def rearrange(arr1,arr2):
  n1=len(arr1)
  n2=len(arr2)
  i,j,k=0,0,0
  arr3=[]
  while i<n1 and j<n2:
    arr3.append(arr1[i])
    arr3.append(arr2[j])
    i+=1
    j+=1

  while i<n1:
    arr3.append(arr1[i])
    i+=1

  while j<n2:
    arr3.append(arr2[j])
    j+=1

  result=''.join(arr3)
  return result
```

```python
def reverse_rearrange(input_string):
  string_to_list=list(input_string)
  list_of_result=[]
  for x in range(len(input_string)):
    new_index=x
    if x!=1 and x!=3 and x!=5:
      list_of_result.insert(new_index,string_to_list[x])
  result=''.join(list_of_result)
  return result
```

```python
# Scheme 1
#hash(rearrangement(password, salt))

import hashlib

def scheme1(password, salt):
    rearranged_string = rearrange(password, salt)
    print(rearranged_string)
    hashed_result = hashlib.md5(rearranged_string.encode()).hexdigest()
    reversed_string = reverse_rearrange(rearranged_string)
    return hashed_result, reversed_string


password = "Secret123"
salt = random_salt(3)
print(salt)
hashed_result, reversed_string = scheme1(password, salt)
print("Hashed Result:", hashed_result)
print("Reversed String:", reversed_string)
```

```python
df_1['temp_result'] = df_1.apply(lambda row: scheme1(row['password'], row['salt']), axis=1)

# Split the temporary column into two separate columns
df_1[['hashed_result', 'reversed_string']] = pd.DataFrame(df_1['temp_result'].tolist(), index=df_1.index)

# Drop the temporary column
#df_1 = df_1.drop('temp_result', axis=1)
```
